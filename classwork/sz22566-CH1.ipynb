```python
import matplotlib.pyplot as plt
import numpy as np

x = range(100)
y = range(100)

plt.subplot(221)
plt.plot(x, np.sin(x))
plt.title("plot 1")

plt.subplot(222)
plt.scatter(x, np.cos(x))
plt.title("plot 2")

plt.subplot(223)
plt.stackplot(x, np.sin(y))
plt.title("plot 3")

plt.subplot(224)
plt.hexbin(x, np.sin(x), gridsize=20)
plt.title("plot 4")

plt.subplots_adjust(left = 1, right = 2, top = 1.5, bottom = 0.5, wspace = 0.5, hspace = 0.5)

```
