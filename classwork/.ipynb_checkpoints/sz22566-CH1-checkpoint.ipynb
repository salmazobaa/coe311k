```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-10,10)
y = np.arange(-10,10)

plt.subplot(2,2,1)
plt.hist(x, 30)
plt.title("plot 1")

plt.subplot(2,2,2)
plt.plot(x, np.cos(x))
plt.title("plot 2")

plt.subplot(2,2,3)
plt.scatter(x, y)
plt.title("plot 3")

plt.subplot(2,2,4)
plt.stackplot(x, np.cos(2*x))
plt.title("plot 4")

plt.subplots_adjust( left = 0, right = 1, bottom = 1, top = 2, wspace = 0.5, hspace = 0.5)
```
