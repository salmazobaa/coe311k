```python
import numpy as np
import math
import matplotlib.pyplot as plt
import random

# Monte Carlo function.

def monte_carlo(num_samples):
    x = 0
    y = 0
    counter = 0
    for i in range(num_samples):
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)
        
        if (x*x + y*y < 1):
            counter += 1
    return 4*counter/ num_samples

# iterating number of samples passed to monte carlo function 
# from 10 to 1000 in steps of 20.

y = []

for i in range(10,1000,20):
    n = round(abs(monte_carlo(i) - np.pi), 15)
    y.append(n)
    
# individually passing 1000 to function because iterative
# loop stops at 990.

l = round(abs(monte_carlo(1000) - np.pi), 15)
y.append(l)

# plotting error.

x = list(range(10, 1001, 20))
x.append(1000)
plt.figure(figsize=(10,6))
plt.plot(x, y)
plt.xlabel("Number of samples", fontsize=16)
plt.ylabel("Absolute error", fontsize=16)
plt.show()

```
