```python
import math
import numpy as np 
import random
import matplotlib.pyplot as plt

#monte carlo function
def approximatePi(numSamples):
    x = 0
    y = 0
    counter = 0
    for i in range(numSamples):
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)
        
        if (x*x + y*y < 1):
            counter = counter + 1
    return 4*counter/ numSamples

# empty lists for plotting later
t = [] 
r = []
for i in range(0,100):
    calculated_pi = approximatePi(1000)
    true_pi = np.pi
    true_error = abs(true_pi-calculated_pi)
    relative_error = abs((true_pi- calculated_pi)/ true_pi)
    
    # populate lists
    t.append(true_error)
    r.append(relative_error)

# plot difference between true error and relative error    
plt.figure()
plt.plot(t)
plt.plot(r)
plt.show()

    
```
